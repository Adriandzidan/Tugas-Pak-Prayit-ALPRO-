# **HASIL PRAKTIKUM**

1.  **Praktikum 01 – Fungsi Dasar dengan def**

```python
def sapa():
  print("Halo, selamat datang di Politeknik Negeri Semarang!")

sapa()
```

3. **Praktikum 03 – Fungsi dengan Default Argument**

```python
def sapa(nama="mahasiswa"):
  print(f"Halo, {nama}!")

sapa()
sapa("Andi")
```

4. **Praktikum 04 – Fungsi dengan *args* dan *kwargs*.**

```python
def jumlahkan (*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print("total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama  ="budi", Umur=20, Prodi="TRK")
```

5. **Praktikum 05 – Scope Variabel**

```python
x = 10

def ubah():
  x = 5
  print ("Nilai lokal:", x)

ubah()
print("Nilai global:", x)
```

6. **Praktikum 06 – Fungsi Lambda**

```python
kali = lambda x, y: x * y
print("Hasil perkalian:", kali(4, 5))
```

7. **Praktikum 07 – Fungsi Decorator**

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan...")
    fungsi()
    print("Setelah fungsi Dijalankan...")
  return wrapper

@decorator_sapa
def sapa():
  print("Halo, selamat datang!")

sapa()
```

8. **Praktikum 08 – Studi Kasus: Menghitung Rata-rata Nilai Mahasiswa**

```python
def hitung_rata(data):
  return sum(data) / len(data)

nilai = [80, 90, 85]
rata = hitung_rata(nilai)

print("Rata-rata nilai:", rata)
```

3. Fungsi dengan Default Argument

```python
def salam(waktu="pagi", nama="mahasiswa") :
  print(f"Selamat {waktu}, {nama}!")

salam()
salam("malam")
salam("sore", "dumbo")
```

4. **Fungsi dengan args**

```python
def jumlahkan_semua(*angka):
  return sum(angka)

print("Total 3 bilangan:", jumlahkan_semua(100, 220, 430))
print("Total 5 bilangan:", jumlahkan_semua(1, 2, 3, 4, 5))
print("Total 7 bilangan:", jumlahkan_semua(100, 15, 20, 70, 40, 50, 60))
```

5. **fungsi kwargs**

```python
def info_barang(**data):
  for key, value in data.items():
    print(f"{key}: {value}")

info_barang(Nama="Buku", Harga=15000, Stok=30)
```

6. **Scope Variabel**

```python
x = 100

def ubah():
  x = 50
  print("Nilai lokal:" , x)

ubah()
print("Nilai Global:" , x)
```

7. **Fungsi Lambda**

```python
kuadrat = lambda x: x ** 2
print("Kuadrat dari 5 adalah:", kuadrat(5))
print("Kuadrat dari 7 adalah:", kuadrat(7))
```

```python
import math
keliling_lingkaran = lambda r: 2 * math.pi * r
print("Keliling lingkaran dengan jari-jari 7 adalah:", keliling_lingkaran(9))
print("Keliling lingkaran dengan jari-jari 10 adalah:", keliling_lingkaran(11))
```

8. **Fungsi Decorator**

```python
def logger(fungsi):
  def wrapper():
    print("Memulai Fungsi....")
    fungsi()
    print("Fungsi Selesai Dijalankan...")
  return wrapper

@logger
def hallo():
  print("Halo")

hallo()
```

9. **Studi Kasus Nyata – Program Nilai**

```python
def input_nilai():
   return [float(x) for x in input("Masukkan nilai mahasiswa (pisahkan dengan spasi): ").split()]

def hitung_rata(data):
  return sum(data) / len(data)

def status_lulus(rata):
  if rata >= 60:
    return "Lulus"
  else:
    return "Tidak Lulus"

nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

nilai = input_nilai()
rata = hitung_rata(nilai)
print("Rata-rata:", rata)
print("Status:", status_lulus(rata))

tugas = float(input("Nilai tugas: "))
ujian = float(input("Nilai ujian: "))
print("Nilai akhir:", nilai_akhir(tugas, ujian))
```
